```python
df['International plan'] = df['International plan'].map({"No":False, "Yes":True})
df.head()
```

```python
df = df.replace({"Voice mail plan":{"No":False, "Yes":True}})
df.head()
```

```python
columns_to_show = ["Total day minutes", "Total eve minutes", "Total night minutes"]
df.groupby(['Churn'])[columns_to_show].describe(percentiles=[])
```

```python
sns.histplot(df['Total day minutes'], kde=True, bins = 30, color = 'blue')
```

```python
sns.histplot(df['Customer service calls'], kde = False, bins=20, color = 'green')
plt.show()
```

```python
sns.countplot(x = 'Churn', data = df, palette='Set2')
plt.show()
```

```python
sns.countplot(x = 'International plan', data=df, palette='pastel')
plt.show()
```

```python
sns.countplot(x = 'Voice mail plan', data = df, palette='muted')
plt.show()
```

```python
sns.boxplot(x ='Churn', y = 'Total day minutes', data = df, palette='Set3')
plt.show()
```

```python
sns.boxplot(x = 'Churn', y = 'Customer service calls', data = df, palette='coolwarm')
plt.show()
```

```python
plt.figure(figsize=(12,8))
corr = df.corr()
sns.heatmap(corr, annot=True, fmt = '.2f', cmap = 'coolwarm')
```

```python
sns.pairplot(df, vars=['Total day minutes', 'Total eve minutes', 'Total night minutes', "Total intl minutes"], hue = 'Churn', palette="husl")
plt.suptitle("Pairplot of Usage Metrics")
plt.show()
```

```python
sns.scatterplot(x = 'Total day minutes', y = 'Total day charge', hue = 'Churn', data=df, palette='viridis')
plt.show()
```

```python
plt.figure(figsize=(12,8))
df['State'].value_counts().plot(kind='bar', color = 'teal')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(data=df[['Total day minutes', 'Total eve minutes', 'Total night minutes']].mean(axis = 0), marker='o')
plt.show()
```

```python
plt.figure(figsize=(10,6))
crosstab = pd.crosstab([df['International plan'],df['Voice mail plan']], df['Churn'])
sns.heatmap(crosstab, annot=True, fmt = 'd', cmap='coolwarm')
plt.show()
```
